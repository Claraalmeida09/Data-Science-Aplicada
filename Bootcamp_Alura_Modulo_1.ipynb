A média é feita apenas com as colunas que possuem todos os dados, o pandas ignora as colunas que possuem algum valor vazio

```python
dados.mean()
```

```python
colunas_usaveis = dados.mean().index
colunas_usaveis
```

Inserindo a coluna que nos informa de qual unidade da Federação é o valor correspondente

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')

colunas_usaveis
```
