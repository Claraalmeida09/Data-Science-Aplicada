```python
import pandas as pd

df = pd.read_csv('/content/A160324189_28_143_208.csv', sep=';', skiprows=3, encoding="ISO-8859-1", skipfooter=12, thousands='.', decimal=',')
df.head(10)
```

Modificando os dados faltantes

```python
pd.options.display.float_format = '{:.2f}'.format 
```

```python
df.replace('-', float('Nan'), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
colunas_usaveis = df.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
```

```python
usaveis = df[colunas_usaveis]
```

```python
usaveis.head()
```

```python
usaveis = usaveis.drop('Total', axis=1)

usaveis.head()
```

```python
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head()
```
